```python
#IMPORT DELLE LIBRERIE
#import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px
import sys, os
from IPython.display import display
```

```python
#IMPORT DEI DATI
sys.argv = "- ./einaudi/gateway/tcp_port_time.dat 172.18.142.254 tcp".split()

if not os.path.exists("images"):
    os.mkdir("images")
    
if len(sys.argv) < 4:
     sys.exit("Wrong Usage, use : python plot_try.py [path_src] [ip_dest] [tcp_or_udp]")

path = sys.argv[1]
display(path)

titleF= sys.argv[2] +'_'+ sys.argv[3]
display(titleF)

df= pd.read_csv(path,delimiter=' ')
display(df)
```

```python
fig = px.scatter(df, x="Time", y="Port",width=600, height=600,
                  labels={"Time": "Time [ms]", "Port": "Port [Number]"}
                )
fig.show()
```

```python
df_group= df.groupby('Port',  as_index=False).count()
display(df_group)
```

```python
fig2 = px.scatter(df_group, x="Port", y="Time", width=1200, height=600,
                  labels={"Time": "Time [ms]", "Port": "Port [Number]"}
                )
fig2.show()
```
